```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data_folder = "./data/flower_data/train/1/"
```

```python
filename = "image_06737.jpg"
```

```python
image = cv2.imread(data_folder+filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
rows,cols,ch = image.shape
```

```python
print(rows)
print(cols)
print(ch)
```

### Affine Transformation

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(image,M,(cols,rows))
```

```python
plt.imshow(dst)
plt.show()
```

```python
plt.subplot(121),plt.imshow(image),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
pts1 = np.float32([[100,100],[495,100],[0,400],[595,400]]) # 1-top left, 2-top right, 3-bottom left, 4-bottom right
pts2 = np.float32([[0,0],[595,0],[0,500],[500,595]])
```

```python
M = cv2.getPerspectiveTransform(pts1,pts2)
```

```python
dst = cv2.warpPerspective(image,M,(595,500)) # width, height
```

```python
plt.subplot(121),plt.imshow(image),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
